```python
import turtle

# Tạo một đối tượng Turtle
pen = turtle.Turtle()

# Thiết lập màu nền
turtle.Screen().bgcolor("pink")

# Vẽ mặt trời
pen.penup()
pen.goto(220, 200)
pen.pendown()
pen.color("yellow")
pen.begin_fill()
pen.circle(50)
pen.end_fill()

# Vẽ tia sáng của mặt trời
pen.penup()
pen.goto(200, 250)
pen.pendown()
pen.color("yellow")
pen.setheading(45)
for _ in range(8):
    pen.forward(80)
    pen.backward(80)
    pen.right(45)
    pen.penup()
    pen.forward(35)
    pen.right(90)
    pen.forward(80)
    pen.pendown()
    pen.backward(80)

# Vẽ căn nhà
pen.penup()
pen.goto(-100, -100)
pen.right(45)
pen.pendown()
pen.color("red")
pen.begin_fill()
for _ in range(4):
    pen.forward(200)
    pen.left(90)
pen.end_fill()

# Vẽ cửa sổ
pen.penup()
pen.goto(-60, -60)
pen.pendown()
pen.color("blue")
pen.begin_fill()
pen.goto(0, -60)
pen.goto(0, 0)
pen.goto(-60, 0)
pen.goto(-60, -60)
pen.end_fill()

#vẽ cửa
pen.penup()
pen.goto(30, -100)
pen.pendown()
pen.begin_fill()
pen.goto(30, -20)
pen.goto(90, -20)
pen.goto(90, -100)
pen.end_fill()

#vẽ mái nhà
pen.penup()
pen.goto (-100,100)
pen.pendown()
pen.color("blue")
pen.goto(0,150)
pen.goto(100,100)
pen.goto(-100,100)
pen.end_fill()

# Vẽ cái cây
pen.penup()
pen.goto(-300, -100)
pen.pendown()
pen.color("brown")
pen.begin_fill()
pen.goto(-260, -100)
pen.goto(-280, 0)
pen.goto(-320, 0)
pen.goto(-340, -100)
pen.end_fill()

pen.penup()
pen.goto(-250, 0)
pen.pendown()
pen.color("green")
pen.begin_fill()
pen.goto(-300, 100)
pen.goto(-350, 0)
pen.goto(-250, 0)
pen.end_fill()

# Đóng cửa sổ Turtle khi bấm chuột vào cửa sổ
turtle.exitonclick()
```
