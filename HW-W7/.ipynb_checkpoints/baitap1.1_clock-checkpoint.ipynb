```python
import turtle  # Sử dụng thư viện Turtle để vẽ đồng hồ

class Clock:
    def __init__(self, hour=0, minute=0, second=0):
        self.hour = hour
        self.minute = minute
        self.second = second
        self.clock_window = turtle.Screen()
        self.clock_window.title("Clock")
        self.pen = turtle.Turtle()
        self.pen.hideturtle()
        self.pen.speed(0)

    def draw_clock_face(self):
        self.pen.penup()
        self.pen.goto(0, -200)
        self.pen.pendown()
        self.pen.circle(200)
        for _ in range(12):
            self.pen.penup()
            self.pen.forward(170)
            self.pen.pendown()
            self.pen.forward(30)
            self.pen.penup()
            self.pen.goto(0, 0)
            self.pen.right(30)

    def draw_clock_hands(self):
        self.pen.penup()
        self.pen.goto(0, 0)
        self.pen.setheading(90)
        angle = (self.hour % 12 + self.minute / 60) * 360 / 12
        self.pen.right(angle)
        self.pen.pendown()
        self.pen.forward(80)

        self.pen.penup()
        self.pen.goto(0, 0)
        self.pen.setheading(90)
        angle = (self.minute + self.second / 60) * 360 / 60
        self.pen.right(angle)
        self.pen.pendown()
        self.pen.forward(120)

        self.pen.penup()
        self.pen.goto(0, 0)
        self.pen.setheading(90)
        angle = self.second * 360 / 60
        self.pen.right(angle)
        self.pen.pendown()
        self.pen.forward(140)

    def update_time(self, hour, minute, second):
        self.hour = hour
        self.minute = minute
        self.second = second

    def run_clock(self):
        self.draw_clock_face()
        while True:
            self.pen.clear()
            self.draw_clock_face()
            self.draw_clock_hands()
            self.clock_window.update()
            self.second += 1
            if self.second == 60:
                self.second = 0
                self.minute += 1
                if self.minute == 60:
                    self.minute = 0
                    self.hour += 1
                    if self.hour == 12:
                        self.hour = 0
            turtle.ontimer(self.run_clock, 1000)

if __name__ == "__main__":
    my_clock = Clock()
    my_clock.run_clock()
    turtle.done()
```
